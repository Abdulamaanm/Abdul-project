```python
df["Quantity Ordered"] = pd.to_numeric(df["Quantity Ordered"], errors='coerce', downcast='integer')
df.isnull().sum()
df.dropna(subset= "Quantity Ordered", inplace= True )
df.isnull().sum()
```

```python
df["Quantity Ordered"]=  df["Quantity Ordered"].astype("int32")
```

```python
# second column
```

```python
df["Price Each"] = pd.to_numeric(df["Price Each"], errors='coerce')
df["Price Each"] = df["Price Each"].astype("float")
df.info()
```

```python
# task 4 add city column
```

```python
df["Sales"]= df["Price Each"] *df["Quantity Ordered"]
df
```

```python
df.dropna(inplace= True)
df.isnull().sum()
```

```python
df["month"]= df["Order Date"].str[0:2]
df["month"]
df
```

```python
group=df.groupby("month")["Sales"].sum()
group
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(group.index, group.values)
plt.xlabel("month")
plt.ylabel("sales")
plt.show()
```

# add time column

```python
df["time"]= df["Order Date"].apply(lambda x: x.split(" ")[1])
df
```

```python
df["hour"]= df["time"].apply(lambda x: x.split(":")[0])
```

```python
c=df.groupby("hour").count()

plt.plot(c.index, c.values, marker="+", color="red")
plt.ylabel("number of order")
plt.grid()
plt.xlabel("time ")
plt.show()
```
